```python
# Объявите функцию encode_df(), которая принимает одну переменную – название csv файла с датасетом. Функция должна:

# Удалить лишние признаки, не влияющие на результат обучения модели.
# Закодировать категориальные признаки с помощью Label Encoder
# Вернуть обработаный датафрейм

from sklearn.preprocessing import LabelEncoder
import pandas as pd


def encode_df(file):
    data = pd.read_csv(file)
    data = data.drop(['PassengerId', 'Name'], axis=1)

    labelencoder = LabelEncoder()
    data['Sex'] = labelencoder.fit_transform(data['Sex'])
    data['Ticket'] = labelencoder.fit_transform(data['Ticket'])
    data['Cabin'] = labelencoder.fit_transform(data['Cabin'])
    data['Embarked'] = labelencoder.fit_transform(data['Embarked'])
    return data
```
