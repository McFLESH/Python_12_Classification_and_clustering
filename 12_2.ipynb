```python
# Объявите функцию train_lr(), которая принимает одну переменную – название csv файла с датасетом. Функция должна:

# Разделить данные на тестовую и тренировочную выборку (при этом за целевой параметр вам необходимо брать значения столбца ‘Survived’) в соотношении 1 к 3.
# Обучить модель логистической регрессии на полученных данных.
# Рассчитать точность модели и вернуть числовое значение в строке return.

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd

def train_lr(file):
    data = pd.read_csv(file)
    features = data.drop('Survived', axis=1)
    target = data['Survived']
    features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.25, random_state=42)

    lreg = LogisticRegression()
    lreg.fit(features_train, target_train)
    lreg_pred = lreg.predict(features_test)

    accuracy = accuracy_score(target_test, lreg_pred)
    return accuracy
print(train_lr('titanic_encoded.csv'))
```
