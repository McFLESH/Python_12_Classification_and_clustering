```python
# Объявите функцию cluster_data(), которая принимает одну переменную – название csv файла с датасетом. Функция должна:

#Сохранить данные из датасета в переменную.
#Обучить модель KMeans на основе полученных данных и распределить данные по кластерам.
#Вернуть результат распределения данных.

from sklearn.cluster import KMeans
import pandas as pd

def cluster_data(file):
    data = pd.read_csv(file)
    kmeans = KMeans(n_clusters=3, random_state=42)
    result = kmeans.fit_predict(data)
    return result
```
