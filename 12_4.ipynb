```python
# Вам требуется создать функцию table_preprocessing, которая примет на вход название CSV-файла для обработки.
#Изучить исходный файл таблицы.
#Удалить один столбец, не имеющий практическую пользу.
#Закодировать столбцы с категориальными переменными, выраженными в строковом формате, в числовой формат.
#Выберите кодировщик, который считаете наиболее подходящим для данного файла (Label Encoder или One-Hot-Encoder).
#Создать переменные X и Y для последующего разделения данных на тренировочную и тестовую выборки,
#где X - это признаки для последующей классификации, а Y - это целевая переменная таблицы.
#Вернуть получившиеся переменные X и Y, указав их через запятую.


from sklearn.preprocessing import LabelEncoder
import pandas as pd

def table_preprocessing(file):
    data = pd.read_csv(file)
    data = data.drop('Booking_ID', axis=1)

    labelencoder = LabelEncoder() #используется LabelEncoder, так как файл большой
    data['type_of_meal_plan'] = labelencoder.fit_transform(data['type_of_meal_plan'])
    data['room_type_reserved'] = labelencoder.fit_transform(data['room_type_reserved'])
    data['market_segment_type'] = labelencoder.fit_transform(data['market_segment_type'])
    data['booking_status'] = labelencoder.fit_transform(data['booking_status'])

    X = data.drop('booking_status', axis=1)
    Y = data['booking_status']
    return X, Y
```
