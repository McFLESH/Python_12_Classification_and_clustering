```python
# Вам требуется создать функцию hotel_booking_analyze, которая примет на вход значение X и Y, которые были получены из предыдущей функции table_preprocessing. Ваши задачи на текущее задание:

#Разделить полученные X и Y на тренировочную и тестовую выборки с помощью метода train_test_split, импортированного из библиотеки sklearn.model_selection. Разделите выборку в соотношении 3 к 1.
#Выбрать три классификатора, обсужденных в лекциях, и для каждого из них выполнить следующие шаги:
#Импортировать классификатор из соответствующей библиотеки (например, from sklearn.svm import SVC).
#Создать объект классификатора (например, svc = SVC()).
#Обучить классификатор с помощью метода fit на тренировочных выборках X_train и Y_train (например, svc.fit(X_train, Y_train)).
#Получить предсказания и записать их в переменную, названную уникально для каждого классификатора (например, y_svc_pred = svc.predict(X_test)).
#Оценить полноту каждого классификатора с помощью метода recall_score библиотеки sklearn.metrics (в программе будет всего 3 вызова этой функции).
#На данном этапе необходимо сравнить полученные в пункте 2.4 переменные с созданной в пункте 1 переменной Y_test (возможно другое название).
#Функция должна вернуть строку "Имя_классификатора_1: полнота_для_классификатора_1. Имя_классификатора_2: полнота_для_классификатора_2. Имя_классификатора_3: полнота_для_классификатора_3."
#Вызовите функцию hotel_booking_analyze с полученными из функции table_preprocessing значениями X и Y и убедитесь, что результат выводится.

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score
import pandas as pd

def table_preprocessing(file):
    data = pd.read_csv(file)
    data = data.drop('Booking_ID', axis=1)

    labelencoder = LabelEncoder() #используется LabelEncoder, так как файл большой
    data['type_of_meal_plan'] = labelencoder.fit_transform(data['type_of_meal_plan'])
    data['room_type_reserved'] = labelencoder.fit_transform(data['room_type_reserved'])
    data['market_segment_type'] = labelencoder.fit_transform(data['market_segment_type'])
    data['booking_status'] = labelencoder.fit_transform(data['booking_status'])

    X = data.drop('booking_status', axis=1)
    Y = data['booking_status']

    return X, Y

def hotel_booking_analyze(x, y):
    X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.25)

    svc = SVC()
    svc.fit(X_train, Y_train)
    Y_svc_pred = svc.predict(X_test)
    recall_svc = recall_score(Y_test, Y_svc_pred)

    k_nbrs = KNeighborsClassifier()
    k_nbrs.fit(X_train, Y_train)
    Y_knbrs_pred = k_nbrs.predict(X_test)
    recall_nbrs = recall_score(Y_test, Y_knbrs_pred)

    lreg = LogisticRegression()
    lreg.fit(X_train, Y_train)
    Y_lreg_pred = lreg.predict(X_test)
    recall_lreg = recall_score(Y_test, Y_lreg_pred)

    return f"Метод опорных векторов: {recall_svc}. Метод ближайших соседей: {recall_nbrs}. Бинарная логистическая регрессия: {recall_lreg}"
X = table_preprocessing('hotel_reservations.csv')[0]
Y = table_preprocessing('hotel_reservations.csv')[1]

result = hotel_booking_analyze(X, Y)
print(result)
```
